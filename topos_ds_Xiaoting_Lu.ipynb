```python
import bs4
from urllib.request import urlopen as uReq
from bs4 import BeautifulSoup as soup
import pandas as pd
import csv
```

```python
my_url = 'https://en.wikipedia.org/wiki/List_of_United_States_cities_by_population'
uClient = uReq(my_url)
page_html = uClient.read()
uClient.close()
page_soup = soup(page_html,'html.parser')
cities = page_soup.find_all("table",class_="wikitable sortable")
cities = cities[0].find_all('tbody')[0]
cities = cities.find_all('tr')[1:]
```

```python
def us_cities():
    us_cities = pd.DataFrame(columns = ['Name','Population_2018','Land_area','Location'])
    
    for i in cities:
        
        rank,city,state,popu2018,popu2010,change,la1,la2,pd1,pd2,location = i.find_all('td')
        # choose city_name, population in 2018, land area and location of the cities
        city = city.find_all('a',href=True)[0]['title']
        popu2018 = popu2018.contents[0]
        la2 = la2.contents[0]
        location = location.find_all('span',class_='geo-dec')[0].contents[0]
        row = {"Name":city,"Population_2018":popu2018,"Land_area":la2,"Location":location}
        us_cities.loc[us_cities.shape[0]+1] = row
    return us_cities
```

```python
us_cities_df = us_cities()
us_cities_df
```

```python
# to csv
us_cities_df.to_csv('us_cities.csv')
```
